```python
matche['home_penalty'].fillna(0, inplace=True)
matche['away_penalty'].fillna(0, inplace=True)
```

```python
matche.head()
```

```python
matche['h_total'] = matche['home_score'] + matche['home_penalty']
matche['a_total'] = matche['away_score'] + matche['away_penalty']
```

```python
matche['h_total'].unique()
```

```python
Runner = matche[matche['Round'] == 'final']
Third = matche[matche['Round'] == 'Third-place match']
```

```python
# Runner up team
def runner_up(row):
  if (row['h_total'] > row['a_total']):
    return row['away_team']
  else:
    return row['home_team']

# third up team
def third_up(row):
  if (row['h_total'] > row['a_total']):
    return row['away_team']
  else:
    return row['home_team']
```

```python
Runner['runner_up'] = Runner.apply(runner_up, axis=1)
Third['Third_place'] = Third.apply(third_up, axis=1)
```

```python
Runner['runner_up'].value_counts().index
```

```python
exp = go.Bar(x=Runner['runner_up'].value_counts().index, y=Runner['runner_up'].value_counts().values ,title='first runner up')
exp.show()
```
